O mais importante na classificação binária é o conjunto de features que iremos avaliar e a classificação que vamos fazer. Por exemplo:

| Imagem de um porco | -> | Feature 1| | Feature 2 | | Feature 3 | -> | Classificação |


```python
'''
Features:
Feature1: tem pelo longo?
Feature2: tem perna curta?
Feature3: late? 
'''
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
# 1 = porco 
# 0 = cachorro
classes = [1, 1, 1, 0, 0, 0]
```

```python
from sklearn.svm import LinearSVC

# cria um modelo linear
model = LinearSVC()
# alimenta com dados e suas classes
model.fit(dados, classes)
```

```python
# cria um novo animal
animalMisterioso = [1, 1, 1]
# pede para o sistema predizer
model.predict([animalMisterioso])
```

```python
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

testes = [misterio1, misterio2, misterio3]

previsoes = model.predict(testes)
```

```python
# respostas reais dos testes
testesClasses = [0, 1, 1]
```

```python
# verifica quantos acertou
corretos = (previsoes == testesClasses).sum()
total = len(testes)
taxaAcerto = corretos/total
print("Taxa de acerto do modelo: ", taxaAcerto * 100)
```
